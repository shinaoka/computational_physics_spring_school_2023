```python
import numpy as np # Linear Algebra
import matplotlib.pyplot as plt # Plotting

N = 100
beta = 1.0 # Inverse temperature

# -2N+1, -2N+3, ..., 2N-3, 2N-1
ns = 2 * np.arange(-N, N) + 1
```

$$
\mathrm{i}\omega_n = \mathrm{i}\frac{(2n+1)\pi}{\beta}
$$

```python
# iw_{-N}, iw_{-N+1}, ..., iw_{N-2}, iw_{N-1}
iws = 1j * ns * np.pi/beta
```

$$
G(\mathrm{i}\omega_n) = \frac{1}{\mathrm{i}\omega_n - \omega_0}
$$

```python
omega0 = 2.0
g = 1/(iws - omega0)
```

```python
plt.plot(iws.imag, g.imag, marker="x")
plt.xlabel("w")
plt.ylabel("Im G(iw)")
```

$$
n =  \frac{1}{1 + e^{\beta \omega_0}}
$$

```python
nref = 1/(1 + np.exp(beta * omega0))
nref
```

$$
\tilde{G}(\iw) \equiv G(\iw) - \frac{1}{\iw} \propto O((1/\iw)^2)
$$

```python
tildeg = g - 1/iws
```

```python
plt.loglog(np.abs(iws.imag), np.abs(tildeg), marker="x")
plt.loglog(np.abs(iws.imag), 1/np.abs(iws)**2, ls="--")
plt.xlabel("|w|")
plt.ylabel("|tilde G(iw)|")
```

$$
n = \frac{1}{\beta} \sum_{n=-N}^{N-1}  \tilde{G}(\iw_n) + \frac{1}{2}.
$$

```python
napprox =  (1/beta) * np.sum(tildeg).real + 0.5
napprox
```

```python
print("error", napprox - nref)
```
